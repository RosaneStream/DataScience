```python
from pickle import TRUE
balance_cc = 0
limit_withdrawn = 500.00
withdraws_num = 0
WITHDRAWS_LIM = 3
statement_cc = " "
value_cc = 0

menu = """
  Please, select an option:

  1 - Statement
  2 - Withdrawn
  3 - Deposit
  4 - Exit """

while True:

  print(menu)
  option_cc = int(input())

  if option_cc == 1:
    print("*********************Bank Statement*********************")
    print(statement_cc)
    print(f'Balance: $ {balance_cc:.2f}')
    print("*********************Bank Statement*********************")
       
  elif option_cc == 2:
    value_cc = float(input("Please, enter the withdrawn value: "))

    if value_cc < 0:
      print("Invalid value!")
      continue
  
    if withdraws_num >= WITHDRAWS_LIM:
        print(f'Sorry! You have permission for just {WITHDRAWS_LIM} withdraws.')

    elif value_cc > limit_withdrawn:
        print(f'Sorry! You are not allowed to withdraw more than {limit_withdrawn:.2f}')

    elif value_cc > balance_cc:
        print("Sorry! You do not have enough balance.")

    else:
        balance_cc -= value_cc
        statement_cc += f'Withdrawn: $ {value_cc:.2f}\n'
        withdraws_num += 1
        print(f'withdraws = {withdraws_num} limite {WITHDRAWS_LIM}')

  elif option_cc == 3:
    value_cc = float(input("Please, enter the deposit value: "))

    if value_cc < 0:
      print("Invalid value!")
      continue

    else:
      balance_cc += value_cc
      statement_cc+= f'Deposit: $ {value_cc:.2f}\n'

  elif option_cc == 4:
    break

  else:
    print("Invalid option! Please, select a valid option!")


```
