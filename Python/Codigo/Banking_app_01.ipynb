```python
import textwrap

### This exercise does not refer deposit or withdraw to some account or client

def menu():
    menu = """\n
    ================ MENU ================
    [1]\tDeposit
    [2]\tWithdraw
    [3]\tStatement
    [4]\tNew client
    [5]\tNew account
    [6]\tList all accounts\n
    
    [7]\tExit
    ================ MENU ================
    => """
    return input(textwrap.dedent(menu))


def deposit(balance_cc, amount_cc, statement_cc, /):
    if amount_cc > 0:
        balance_cc += amount_cc
        statement_cc += f"Deposit: $ {amount_cc:.2f}\n"
        print(f"\n=== Deposit confirmed with success: $ {amount_cc:.2f}===")
        print(f"Balance ==> $ {balance_cc:.2f}")
    else:
        print("\n@@@ Deposit has been failed! @@@")

    return balance_cc, statement_cc


def withdraw(*, balance_cc, amount_cc, statement_cc, withdraw_limit, withdraws_num, withdraw_max):
   
    if withdraws_num >= withdraw_max:
        print("\n@@@ Transaction has been failed! Maximum number of withdrawals exceeded. @@@") 
    
    elif amount_cc > withdraw_limit:
        print("\n@@@ Transaction has been failed! Withdrawal exceeds the amount limit. @@@")

    elif amount_cc > balance_cc:
        print("\n@@@ Transaction has been failed! You have no enough balance. @@@")

    elif amount_cc > 0:
        balance_cc -= amount_cc
        statement_cc += f"withdraw: $ {amount_cc:.2f}\n"
        withdraws_num += 1
        print(f"\n=== withdraw confirmed with success: $ {amount_cc:.2f}===")
        print(f"Balance ==> $ {balance_cc:.2f}")

    else:
        print("\n@@@ withdraw has been failed! @@@")

    return balance_cc, statement_cc, withdraws_num


def show_statement(balance_cc, /, *, statement_cc):
    print("*********************Bank Statement*********************")
    print(statement_cc)
    print(f'Balance: $ {balance_cc:.2f}')
    print("*********************Bank Statement*********************")


def create_client(clients):
    document = input("Enter the number of your document (only numbers): ")
    client = search_client(document, clients)

    if client:
        print("\n@@@ Client is already registered! @@@")
        return

    name = input("Enter the full name: ")
    birth_date = input("Enter the birth date (dd-mm-aaaa): ")
    address = input("Enter the address (street, number - City/Two-letter State Abbr): ")

    clients.append({"name": name, "birth_date": birth_date, "document": document, "address": address})

    print("=== The client has been successfully registered! ===")


def search_client(document, clients):
    clients_found = [client for client in clients if client["document"] == document]
    return clients_found[0] if clients_found else None


def create_account(agency, account_num, clients, accounts):
    document = input("Enter the document number (only numbers): ")
    client = search_client(document, clients)

    if client:
        account =  {"agency": agency, "account_num": account_num, "client": client}
        accounts.append(account)
        return account

    print("\n@@@ Client invalid. Transaction aborted! @@@")


def show_accounts(accounts):
    for account in accounts:
        text_line = f"""\
            Agency:\t{account['agency']}
            Account:\t{account['account_num']}
            Client:\t{account['client']['name']}
        """
        print("=" * 100)
        print(textwrap.dedent(text_line))


def main():

    balance_cc = 0
    withdraw_limit = 500.00
    withdraws_num = 0
    statement_cc = " "
    amount_cc = 0
    account_add = 1

    WITHDRAW_MAX = 3
    AGENCY = "0001"

    clients = []
    accounts = []

    while True:
        option = menu()

        if option == "1":
            amount_cc = float(input("Please, enter the deposit amount: "))

            balance_cc, statement_cc = deposit(balance_cc, amount_cc, statement_cc)

        elif option == "2":
            amount_cc = float(input("Please, enter the withdraw amount: "))

            balance_cc, statement_cc, withdraws_num = withdraw(
                balance_cc = balance_cc,
                amount_cc = amount_cc,
                statement_cc = statement_cc,
                withdraw_limit = withdraw_limit,
                withdraws_num = withdraws_num,
                withdraw_max = WITHDRAW_MAX
            )   

        elif option == "3":# statement
            show_statement(balance_cc, statement_cc = statement_cc)

        elif option == "4": # new client
            create_client(clients)

        elif option == "5": # new account
            #account_num = len(accounts) + 1
            account_num = account_add
            account = create_account(AGENCY, account_num, clients, accounts)

            if account:
                account_add += account_add
                print("=== Account has been successfully created! ===")

        elif option == "6":
            show_accounts(accounts)

        elif option == "7":
            break

        else:
            print("@@@ Invalid option! Please, select a valid option! @@@")


main()
```
